# The Most Aesthetically Pleasing Neighborhood

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
trees = pd.read_csv("trees.csv", low_memory = False)
```

```python
values = trees['neighborhood'].value_counts()
valueshigh = values.iloc[0:5]
```

## Neighborhoods with the Most Trees

```python
valueshigh.plot.bar(x = values[0], y = values[1])
```

## Neighborhoods with the Least Trees

```python
valueslow = values.iloc[85:90]
valueslow.plot.bar(x = valueslow[0], y = valueslow[1])
```
